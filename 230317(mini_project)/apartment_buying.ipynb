```python
import pandas as pd
import numpy as np
import seaborn as sns
import folium
import json
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
# 파일 불러오기

df2 = pd.read_excel("C:/python/datas/아파트(매매)_실거래가.xlsx")
df2
```

```python
# 불필요한 열 삭제

df2.drop(df2.index[0:16], inplace=True)
df2
```

```python
# 컬럼 이름 바꾸기 및 불필요한 행 삭제

df2.rename(columns = {'□ 본 서비스에서 제공하는 정보는 법적인 효력이 없으므로 참고용으로만 활용하시기 바랍니다.': '시군구', 
                     'Unnamed: 8': '거래금액(만원)',
                     'Unnamed: 5' : '전용면적(㎡)'}, inplace=True)
del df2['Unnamed: 1']
del df2['Unnamed: 2']
del df2['Unnamed: 3']
del df2['Unnamed: 4']
del df2['Unnamed: 6']
del df2['Unnamed: 7']
del df2['Unnamed: 9']
del df2['Unnamed: 10']
del df2['Unnamed: 11']
del df2['Unnamed: 12']
del df2['Unnamed: 13']
del df2['Unnamed: 14']
df2
```

```python
# dtype을 object에서 float64로 변경

df2["전용면적(㎡)"] = df2['전용면적(㎡)'].astype('float64')
```

```python
# 따옴표와 공백제거

df2['거래금액(만원)'] = df2['거래금액(만원)'].str.strip().replace("'", '')
```

```python
# 숫자들 사이에 있는 콤마를 제거하고 dtype을 float으로 변경

df2['거래금액(만원)'] = df2['거래금액(만원)'].str.replace(",", '').astype('f')
```

```python
# 전용면적 당 거래금액(㎡)을 구하고 그 결과값을 소수점 2번째까지 표현

df2['전용면적 당 거래금액(㎡)'] = round((df2['거래금액(만원)']) / (df2['전용면적(㎡)']),2)
```

```python
df2
```

```python
# '시군구' 문자열 컬럼을 분할하여 '구'를 인덱싱

df2.시군구 = df2.시군구.str.split().str[1]
df2
```

```python
# 그룹별로 데이터를 집계 후 열 평균을 구하고 그 결과값의 소수점 2번째까지 표현

df3 = round(df2.groupby('시군구').mean(), 2)
df3
```

```python
# '거래금액(만원')의 dtype을 float32에서 float64로 변경

df3['거래금액(만원)'] = df3['거래금액(만원)'].astype('float64')
df3
```

```python
# df3의 값을 소수점 2번째까지 표현

df3 = round(df3, 2)
df3
```

```python
# '시군구' index를 컬럼자리로 이동

df3 = df3.reset_index()
```

```python
# 지도 상 "전용면적 당 거래금액(㎡)"에 대해 금액 크기에 따라서 색깔로 구분 

seoul_geojson = "https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json"

m = folium.Map(location = [37.57, 126.99], zoom_start = 11)
folium.Choropleth(geo_data = seoul_geojson,
                  data=df3[["시군구", "전용면적 당 거래금액(㎡)"]],
                  columns=["시군구", "전용면적 당 거래금액(㎡)"],
                  key_on="feature.properties.name",
                  fill_color = "PuRd",
                  fill_opacity = 0.4,
                  line_opacity = 1
).add_to(m)
```

```python
# 전용면적 당 거래금액(㎡)에 가독성 확보를 위해 콤마 삽입

df3['전용면적 당 거래금액(㎡)'] = df3['전용면적 당 거래금액(㎡)'].apply(lambda int_num : '{:,}'.format(int_num))
df3
```
