```python
# DataFrame의 값 정렬하기

## sort_values() 메서드

titanic.sort_values("fare") # "fare" 컬럼에 대해 값 정렬(오름차순)
```

```python
# 첫 번째 인자는 by 키워드를 가지며 정렬할 칼럼을 지정 가능
titanic.sort_values(by = ["fare", "age"]) 
# "fare", "age" 컬럼에 대해 값 정렬(오름차순)
```

```python
# 내림차순을 보고 싶다면 ascending = Fasle 설정
titanic.sort_values(by = ["fare", "age"], ascending =  False) 
```

```python
# 각 컬럼마다 정렬을 달히하고 싶으면 ascending = [True, False, False]
# 리스트 형태로 컬럼명 전달 (True는 오름차순, False는 내림차순)

# na_position = "first", "last"를 통해 결측치 정렬 방향 지정 가능

titanic.sort_values(by = ["fare", "age", "sex"], 
                   ascending = [False, True, True],
                   na_position = "last"
                  )
```

```python
# DataFrame 고급 인덱싱

# loc : label 기반의 2차원 인덱싱
# df.loc[row 인덱싱 값]
# df.loc[row 인덱싱 값, column 인덱싱 값]

# iloc : 순서를 나타내는 정수 기반의 2차원 인덱싱
# df.iloc[0]
# df.iloc[0,1], df.iloc[:2. -1], ...

## loc 인덱서 
```

```python
# row 인덱싱 값은 정수 또는 정수 index데이터이고,
# column 인덱싱 값은 label 문자열

df = pd.DataFrame(np.arange(10, 22).reshape(3, 4),
                  index = ['a', 'b', 'c'],
                  columns = ['A', 'B', 'C', 'D'])
df
```

```python
# 인덱스를 하나만 넣으면 row 선택
# 선택된 row를 Series로 반환
df.loc["a"]
```

```python
# 슬라이싱도 가능하지만 loc을 안 쓰는 것과 결과는 동일
df.loc["b":"c"]
```

```python
df["b":"c"]
```

```python
### 인덱스 데이터의 리스트 인덱싱

df.loc[["b","c"]]
#  loc을 안 쓰면 KeyError 발생
```

```python
df.A > 15 # 영어 문자열을 속성처럼 접근 가능
```

```python
df.loc[df.A>15]
```

```python
### callable한 함수의 인덱싱

def select_rows(df, num):
    return df.A > num
```

```python
select_rows(df, 10)
```

```python
df.loc[select_rows(df, 10)]
```

```python
## loc 인덱서의 주의사항

# loc 인덱서는 column label 인덱싱이나
# column label 리스트 인덱싱은 불가능

# df.loc["A"] # KeyError 발생

# df.loc[["A", "B"]] # KeyError 발셍

# 원래 row index 값이 정수인 경우 마지막 값이 포함됨
df2 = pd.DataFrame(np.arange(10, 26).reshape(4, 4),
                   columns = ["A", "B", "C", "D"])
df2
```

```python
df2.loc[1:2] # 끝값 2가 포함됨
# 원래 정수형 index이므로 끝값 포함
```

```python
# loc과 iloc의 차이
# iloc은 label 인덱스가 아닌 숫자로된 인덱스로 접근하기
# 때문에 끝값 포함이 안 됨

df2.loc[1:2]
```

```python
df2.iloc[1:2]
```

```python
## row와 column의 loc 인덱서

df.loc["a", "A"]
# row index 값이 a, label 값이 A인 위치의 결과
```

```python
# 슬라이싱
df.loc["b":,"A"]
```

```python
df.loc["a", :]
```

```python
df.loc[["a", "b"], ["B", "D"]]
```

```python
### Boolean indexing

df.loc[df.A > 10, ["C", "D"]]
```

```python
## iloc 인덱서

# iloc 인덱서는 loc 인덱서와 달리 label이 아닌 정수 인덱스만 받는다.
# 다른 부분은 loc 인덱서와 동일

df.iloc[0, 1]
```

```python
df.iloc[:2, 2]
```

```python
df.iloc[0, -2:]
```

```python
df.iloc[2:3, 1:3]
```

```python
# 인덱스가 하나만 들어가면 loc과 마찬가지로 행을 선택
df.iloc[-1]
```
