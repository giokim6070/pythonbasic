```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
```

```python
fuel = sns.load_dataset("mpg")
fuel.head()

# 참고로 mpg는 자동차 연비에 관한 데이터이다..
```

```python
fuel.info()
```

```python
sns.histplot(data = fuel, x = 'horsepower')

# 전체 데이터 속 차랑들의 마력을 파악하기 위해서 fuel 데이터의 마력 분포를 출력했다.
```

```python
sns.jointplot(data = fuel, x = 'horsepower', y = 'mpg', hue = 'origin')

# fuel데이터 속의 생산국에 따른 연비와 마력의 분포를 시각적으로 확인할 수 있게 되었다.
```

```python
sns.swarmplot(
    data = fuel, x = 'origin', y = 'mpg',
    marker = 'x', linewidth = 2
)

# swarmplot()의 사용으로 내가 원하는 시각적 효과를 적용시켰다. 
# 그 결과 국가별 마력, 연비 등의 요소를 더 알아보기 쉽게 파악할 수 있었다.
```

```python
sns.swarmplot(
    data = fuel, x = 'origin', y = 'horsepower',
    marker = 'x', linewidth = 2
)
```
